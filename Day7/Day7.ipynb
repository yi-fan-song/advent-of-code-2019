```python
class UnknownInstruction(Exception):
    pass

class ProgramExit(Exception):
    pass
```

```python
def getParamMode(s, n):
    res = [0,] * n
    if len(s) >= 3:
        res[0] = int(s[-3])
    if len(s) >= 4:
        res[1] = int(s[-4])
    if len(s) >= 5:
        res[2] = int(s[-5])
    return res

def intCodeRun(c, phase, inp):
    i = 0
    inputNo = 0
    while len(c[i]) < 2 or int(c[i][-2:]) != 99:
        if len(c[i]) < 2:
            c[i] = "0" + c[i]
            
        # Add 1 and 2 set at 3
        if int(c[i][-2:]) == 1:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            c[int(param[2])] = str(int(param[0]) + int(param[1]))
            i = i + 4 
            continue
            
        # Multiply 1 and 2 set at 3
        if int(c[i][-2:]) == 2:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            c[int(param[2])] = str(int(param[0]) * int(param[1]))
            i = i + 4 
            continue
            
        # Ask for input set at 1
        if int(c[i][-2:]) == 3:
            m = getParamMode(c[i], 1)
            #inpt = input("Please enter an input: ")
            if inputNo == 0:
                inpt = phase
            if inputNo == 1:
                inpt = inp
            if inputNo == 2:
                print("too many inputs needed")
            inputNo += 1
            c[int(c[i + 1])] = inpt
            i = i + 2 
            continue
            
        # Print 1
        if int(c[i][-2:]) == 4:
            m = getParamMode(c[i], 1)
            out = c[i + 1]
            if m[0] == 0:
                out = c[int(out)]
            #print("output:", out)
            return out
            #i = i + 2 
            #continue
                
        # Jump to 2 if 1 is not false
        if int(c[i][-2:]) == 5:
            m = getParamMode(c[i], 2)
            param = [c[i+1], c[i+2]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            if int(param[0]) != 0:
                i = int(param[1])
                continue
            i = i + 3
            continue
            
        # Jump to 2 if 1 is false
        if int(c[i][-2:]) == 6:
            m = getParamMode(c[i], 2)
            param = [c[i+1], c[i+2]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            if int(param[0]) == 0:
                i = int(param[1])
                continue
            i = i + 3
            continue
            
        # 1 less than 2 set at 3
        if int(c[i][-2:]) == 7:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            c[int(param[2])] = str(int(int(param[0]) < int(param[1])))
            i = i + 4 
            continue
            
        # 1 equals 2 set at 3
        if int(c[i][-2:]) == 8:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            c[int(param[2])] = str(int(int(param[0]) == int(param[1])))
            i = i + 4 
            continue
            
        print("unknown instruction", c[i], i)
        break
```

```python
file = open("input.txt", "r")

content = file.read()
content = content.split(',')
content = [each.replace('\n', '') for each in content]

copy = content.copy()

#print(content)
```

```python
def runAmp(c, copy, phase):
    output = 0
    for i in range(5):
        c = copy.copy()
        output = intCodeRun(content, phase[i], output)
    return int(output)
```

```python
combinations = itertools.permutations((0,1,2,3,4))
```

```python
content = copy.copy()
best = 0
bestPhase = (0, 0, 0, 0, 0)

for c in combinations:
    res = runAmp(content, copy, c)
    if best < res:
        best = res
        bestPhase = c
```

```python
print("part1: ", best, bestPhase)
```

```python
def intCodeRun(c, phase, inp, skipPhase = False, initPointer = 0):
    i = initPointer
    inputNo = 0
    while len(c[i]) < 2 or int(c[i][-2:]) != 99:
        if len(c[i]) < 2:
            c[i] = "0" + c[i]
        #print("operation: ", c[i])
        
        # Add 1 and 2 set at 3
        if int(c[i][-2:]) == 1:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            #print("paramMode: ", m, " params: ", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            #print("finalParam: ", param)
            c[int(param[2])] = str(int(param[0]) + int(param[1]))
            #print("result:", c[int(param[2])])
            i = i + 4 
            continue
            
        # Multiply 1 and 2 set at 3
        if int(c[i][-2:]) == 2:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            #print("paramMode: ", m, " params: ", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            #print("finalParam: ", param)
            c[int(param[2])] = str(int(param[0]) * int(param[1]))
            #print("result:", c[int(param[2])])
            i = i + 4 
            continue
            
        # Ask for input set at 1
        if int(c[i][-2:]) == 3:
            m = getParamMode(c[i], 1)
            #inpt = input("Please enter an input: ")
            if inputNo == 0:
                if skipPhase:
                    #print("skipping phase, input: ", inp)
                    inpt = inp
                    inputNo += 1
                else:
                    #print("phase: ", phase)
                    inpt = phase
            elif inputNo == 1:
                #print("input: ", inp)
                inpt = inp
            elif inputNo == 2:
                print("too many inputs needed")
            inputNo += 1
            c[int(c[i + 1])] = inpt
            i = i + 2 
            continue
            
        # Print 1
        if int(c[i][-2:]) == 4:
            m = getParamMode(c[i], 1)
            out = c[i + 1]
            if m[0] == 0:
                out = c[int(out)]
            #print("output:", out)
            i = i + 2 
            return (out, i)
            #continue
                
        # Jump to 2 if 1 is not false
        if int(c[i][-2:]) == 5:
            m = getParamMode(c[i], 2)
            param = [c[i+1], c[i+2]]
            #print("jump if true:", param, m)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            #print("finalParam: ", param)
            if int(param[0]) != 0:
                #print("jumping")
                i = int(param[1])
                continue
            i = i + 3
            continue
            
        # Jump to 2 if 1 is false
        if int(c[i][-2:]) == 6:
            m = getParamMode(c[i], 2)
            param = [c[i+1], c[i+2]]
            #print("jump if false:", param, m)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            #print("finalParam: ", param)
            if int(param[0]) == 0:
                #print("jumping")
                i = int(param[1])
                continue
            i = i + 3
            continue
            
        # 1 less than 2 set at 3
        if int(c[i][-2:]) == 7:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            #print("paramMode: ", m, " params: ", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            #print("finalParam: ", param)
            c[int(param[2])] = str(int(int(param[0]) < int(param[1])))
            #print("result:", c[int(param[2])])
            i = i + 4 
            continue
            
        # 1 equals 2 set at 3
        if int(c[i][-2:]) == 8:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            #print("paramMode: ", m, " params: ", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            #print("finalParam: ", param)
            c[int(param[2])] = str(int(int(param[0]) == int(param[1])))
            #print("result:", c[int(param[2])])
            i = i + 4 
            continue
            
        print("unknown instruction", c[i], i)
        print("code state:", c)
        raise UnknownInstruction()
    raise ProgramExit()
```

```python
def feedback(copy, phase):
    output = 0
    code = [copy.copy(), copy.copy(), copy.copy(), copy.copy(), copy.copy()]
    pointer = [0, 0, 0, 0, 0]
    #print(code[0])
    for i in range(5):
        output, pointer[i] = intCodeRun(code[i], phase[i], output)
    try:
        while True:
            #print(code[0])
            for i in range(5):
                output, pointer[i]  = intCodeRun(code[i], phase[i], output, True, pointer[i])
    except ProgramExit:
        return int(output)
```

```python
best = 0
bestPhase = (0, 0, 0, 0, 0)

phases = itertools.permutations((5,6,7,8,9))
for phase in phases:
    print("trying phase: ", phase)
    ret = feedback(copy, phase)
    if ret is not None and ret > best:
        best = ret
        bestPhase = phase
        
print(best, bestPhase)
```
