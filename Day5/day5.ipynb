```python
def getParamMode(s, n):
    res = [0,] * n
    if len(s) >= 3:
        res[0] = int(s[-3])
    if len(s) >= 4:
        res[1] = int(s[-4])
    if len(s) >= 5:
        res[2] = int(s[-5])
    return res

def intCodeRun(c):
    i = 0
    while len(c[i]) < 2 or int(c[i][-2:]) != 99:
        if len(c[i]) < 2:
            c[i] = "0" + c[i]
            
        # Add 1 and 2 set at 3
        if int(c[i][-2:]) == 1:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            c[int(param[2])] = str(int(param[0]) + int(param[1]))
            i = i + 4 
            continue
            
        # Multiply 1 and 2 set at 3
        if int(c[i][-2:]) == 2:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            c[int(param[2])] = str(int(param[0]) * int(param[1]))
            i = i + 4 
            continue
            
        # Ask for input set at 1
        if int(c[i][-2:]) == 3:
            m = getParamMode(c[i], 1)
            inpt = input("Please enter an input: ")
            c[int(c[i + 1])] = inpt
            i = i + 2 
            continue
            
        # Print 1
        if int(c[i][-2:]) == 4:
            m = getParamMode(c[i], 1)
            out = c[i + 1]
            if m[0] == 0:
                out = c[int(out)]
            print("output:", out)
            i = i + 2 
            continue
                
        # Jump to 2 if 1 is not false
        if int(c[i][-2:]) == 5:
            m = getParamMode(c[i], 2)
            param = [c[i+1], c[i+2]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            if int(param[0]) != 0:
                i = int(param[1])
                continue
            i = i + 3
            continue
            
        # Jump to 2 if 1 is false
        if int(c[i][-2:]) == 6:
            m = getParamMode(c[i], 2)
            param = [c[i+1], c[i+2]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            if int(param[0]) == 0:
                i = int(param[1])
                continue
            i = i + 3
            continue
            
        # 1 less than 2 set at 3
        if int(c[i][-2:]) == 7:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            c[int(param[2])] = str(int(int(param[0]) < int(param[1])))
            i = i + 4 
            continue
            
        # 1 equals 2 set at 3
        if int(c[i][-2:]) == 8:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            c[int(param[2])] = str(int(int(param[0]) == int(param[1])))
            i = i + 4 
            continue
            
        print("unknown instruction", c[i], i)
        break
        

def intCodeRunVerbose(c):
    i = 0
    while len(c[i]) < 2 or int(c[i][-2:]) != 99:
        if len(c[i]) < 2:
            c[i] = "0" + c[i]
        print("operation: ", c[i])
            
        # Add 1 and 2 set at 3
        if int(c[i][-2:]) == 1:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            print("paramMode: ", m, " params: ", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            print("finalParam: ", param)
            c[int(param[2])] = str(int(param[0]) + int(param[1]))
            print("result:", c[int(param[2])])
            i = i + 4 
            continue
            
        # Multiply 1 and 2 set at 3
        if int(c[i][-2:]) == 2:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            print("paramMode: ", m, " params: ", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            print("finalParam: ", param)
            c[int(param[2])] = str(int(param[0]) * int(param[1]))
            print("result:", c[int(param[2])])
            i = i + 4 
            continue
            
        # Ask for input set at 1
        if int(c[i][-2:]) == 3:
            m = getParamMode(c[i], 1)
            inpt = input("Please enter an input: ")
            c[int(c[i + 1])] = inpt
            i = i + 2 
            continue
            
        # Print 1
        if int(c[i][-2:]) == 4:
            m = getParamMode(c[i], 1)
            out = c[i + 1]
            if m[0] == 0:
                out = c[int(out)]
            print("index:", i, "index of output:", c[i + 1])
            print("output:", out)
            i = i + 2
            continue
            
        # Jump to 2 if 1 is not false
        if int(c[i][-2:]) == 5:
            m = getParamMode(c[i], 2)
            param = [c[i+1], c[i+2]]
            print("Param:", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            print("finalParam: ", param)
            if int(param[0]) != 0:
                i = int(param[1])
                print("Jumped to", i)
                continue
            i = i + 3
            continue
            
        # Jump to 2 if 1 is false
        if int(c[i][-2:]) == 6:
            m = getParamMode(c[i], 2)
            param = [c[i+1], c[i+2]]
            print("Param:", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            print("finalParam: ", param)
            if int(param[0]) == 0:
                i = int(param[1])
                print("Jumped to", i)
                continue
            i = i + 3
            continue
            
        # 1 less than 2 set at 3
        if int(c[i][-2:]) == 7:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            print("paramMode: ", m, " params: ", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            print("finalParam: ", param)
            c[int(param[2])] = str(int(int(param[0]) < int(param[1])))
            print("result:", c[int(param[2])])
            i = i + 4 
            continue
            
        # 1 equals 2 set at 3
        if int(c[i][-2:]) == 8:
            m = getParamMode(c[i], 3)
            param = [c[i+1], c[i+2], c[i+3]]
            print("paramMode: ", m, " params: ", param)
            if m[0] == 0:
                param[0] = c[int(param[0])]
            if m[1] == 0:
                param[1] = c[int(param[1])]
            print("finalParam: ", param)
            c[int(param[2])] = str(int(int(param[0]) == int(param[1])))
            print("result:", c[int(param[2])])
            i = i + 4 
            continue
            
        print("unknown instruction", c[i], i)
        break
```
