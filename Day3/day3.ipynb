```python
import math
```

```python
file = open("input.txt", "r")

fileLines = file.readlines()
print(fileLines)
```

```python
fileLines = [[item.replace('\n', '') for item in line.split(',')] for line in fileLines]
print(fileLines)
```

```python
path1 = [[0,0]]
path2 = [[0,0]]

for route in fileLines[0]:
    last = path1[-1]
    if route[0] == 'L':
        for i in range(1, int(route[1:]) + 1):
            path1.append([last[0] - i, last[1]])
    if route[0] == 'R':
        for i in range(1, int(route[1:]) + 1):
            path1.append([last[0] + i, last[1]])
    if route[0] == 'D':
        for i in range(1, int(route[1:]) + 1):
            path1.append([last[0], last[1] - i])
    if route[0] == 'U':
        for i in range(1, int(route[1:]) + 1):
            path1.append([last[0], last[1] + i])
            
            
for route in fileLines[1]:
    last = path2[-1]
    if route[0] == 'L':
        for i in range(1, int(route[1:]) + 1):
            path2.append([last[0] - i, last[1]])
    if route[0] == 'R':
        for i in range(1, int(route[1:]) + 1):
            path2.append([last[0] + i, last[1]])
    if route[0] == 'D':
        for i in range(1, int(route[1:]) + 1):
            path2.append([last[0], last[1] - i])
    if route[0] == 'U':
        for i in range(1, int(route[1:]) + 1):
            path2.append([last[0], last[1] + i])
            
```

```python
def intersectOfCoords(lst1, lst2):
    return set(map(tuple, lst1)) & set(map(tuple, lst2)) 
```

```python
intersect = intersectOfCoords(path1, path2)
```

```python
print(intersect)
```

```python
def manhatDist(tupl):
    return abs(tupl[0]) + abs(tupl[1])

intersect.remove((0,0))

manhattanDistanceList = set(map(manhatDist, intersect))

print(intersect)
min(manhattanDistanceList)
```

```python
closest = 9999999999
closestTuple = (0,0)

for x in intersect:
    dst1 = path1.index(list(x))
    dst2 = path2.index(list(x))
    distance = dst1 + dst2
    print("total:", distance, ", distance 1:", dst1, ", distance 2:", dst2)
    print(path1[dst1], ",", path2[dst2])
    if distance <= closest:
        closest = distance
        closestTuple = x

print(closestTuple, closest)
```
